```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None  # Tambahan untuk Double Linked List

class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    # Fungsi cetak list (maju)
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval, end=" <-> ")
            printval = printval.nextval
        print("None")

    # Fungsi sisip di tengah (setelah node tertentu)
    def AddInBetween(self, middle_node_key, newdata):
        curr = self.headval
        # Cari node tengahnya dulu
        while curr is not None:
            if curr.dataval == middle_node_key:
                break
            curr = curr.nextval

        if curr is None:
            print("Node tidak ditemukan")
            return

        NewNode = Node(newdata)

        # Atur pointer NewNode
        NewNode.nextval = curr.nextval
        NewNode.prevval = curr

        # Atur pointer node setelahnya (jika ada) agar mundur ke NewNode
        if curr.nextval is not None:
            curr.nextval.prevval = NewNode

        # Atur pointer node saat ini agar maju ke NewNode
        curr.nextval = NewNode

    # Fungsi hapus dengan kata kunci
    def RemoveByKey(self, key):
        curr = self.headval

        # Cek jika head yang mau dihapus
        if curr is not None and curr.dataval == key:
            self.headval = curr.nextval
            if self.headval is not None:
                self.headval.prevval = None
            curr = None
            return

        # Cari node yang mau dihapus
        while curr is not None:
            if curr.dataval == key:
                break
            curr = curr.nextval

        if curr is None:
            print("Data tidak ada")
            return

        # Ubah link nextval dan prevval node tetangga
        if curr.nextval is not None:
            curr.nextval.prevval = curr.prevval

        if curr.prevval is not None:
            curr.prevval.nextval = curr.nextval

        curr = None

# --- EKSEKUSI ---
DLL = DoubleLinkedList()

# Buat 3 Node Manual sesuai gaya modul
n1 = Node("Senin")
n2 = Node("Selasa")
n3 = Node("Rabu")

# Sambungkan Node (Double Link)
DLL.headval = n1
n1.nextval = n2
n2.prevval = n1
n2.nextval = n3
n3.prevval = n2

print("Awal:")
DLL.listprint()

# Lakukan Penyisipan Tengah (Misal: 'Libur' setelah 'Senin')
print("\nSetelah sisip tengah:")
DLL.AddInBetween("Senin", "Libur")
DLL.listprint()

# Lakukan Hapus dengan kata kunci (Misal: hapus 'Selasa')
print("\nSetelah hapus kata kunci 'Selasa':")
DLL.RemoveByKey("Selasa")
DLL.listprint()
```

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Fungsi cetak (perlu break biar nggak looping selamanya)
    def listprint(self):
        printval = self.headval
        if self.headval is not None:
            while True:
                print(printval.dataval, end=" -> ")
                printval = printval.nextval
                if printval == self.headval:
                    break
            print("(Balik ke Head)")
        else:
            print("List Kosong")

    # Fungsi sisip di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        # Cari node terakhir (yang next-nya ke head)
        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        # Ubah link
        last.nextval = NewNode
        NewNode.nextval = self.headval

    # Fungsi hapus di akhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        # Jika cuma ada 1 node
        if self.headval.nextval == self.headval:
            self.headval = None
            return

        # Cari node sebelum terakhir (second to last)
        last = self.headval
        while last.nextval.nextval != self.headval:
            last = last.nextval

        # Putus link node terakhir, arahkan ke head
        last.nextval = self.headval

# --- EKSEKUSI ---
CLL = CircularLinkedList()

# Buat 3 Node Manual
n1 = Node("Pagi")
n2 = Node("Siang")
n3 = Node("Malam")

# Link Circular (n3 balik ke n1)
CLL.headval = n1
n1.nextval = n2
n2.nextval = n3
n3.nextval = n1

print("Awal:")
CLL.listprint()

# Penyisipan Akhir
print("\nSetelah sisip akhir ('Subuh'):")
CLL.AddInEnd("Subuh")
CLL.listprint()

# Hapus Akhir
print("\nSetelah hapus akhir:")
CLL.RemoveEnd()
CLL.listprint()
```
